# 3.4 Regular expressions for detecting word patterns

```python
nltk.download('words')
```

```python
import re
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
```

```python
[w for w in wordlist if re.search(r'ed$', w)]
```

```python
[w for w in wordlist if re.search(r'^..j..t..$', w)]
```

```python
nltk.download('treebank')
```

```python
wsg = sorted(set(nltk.corpus.treebank.words()))
```

```python
set1 = set(w for w in wsg if re.search('ed|ing$', w))
set2 = set(w for w in wsg if re.search('(ed|ing)$', w))
```

```python
[vs for word in wsg for vs in re.findall(r'[aeiou]{2,}', word)]
```

```python
[n for n in re.findall(r'(\d{4})-(\d{2})-(\d{2})', '2009-12-31ab2010-08-19')]
```

```python
nltk.download('toolbox')
```

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
```

```python
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]
cvs
```

```python
cfd = nltk.ConditionalFreqDist(cvs)
```

```python
cfd.tabulate()
```

```python
cv_index = nltk.Index((cv, w) for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w))
```

```python
cv_index['su']
```

```python
cv_index['po']
```
